# fetch all available questions


API doc: https://opentdb.com/api_config.php

```python
GET_TOKEN_URL = "https://opentdb.com/api_token.php?command=request"
res = requests.get(GET_TOKEN_URL).json()
token = res["token"]
token
```

```python
questions = []

while True:
    print("")
    URL = "https://opentdb.com/api.php?amount=50&type=multiple&token="+token
    res = requests.get(URL).json()
    if(res["response_code"] !=0):
        break
    new_questions = res["results"]
    print(f"fetched new questions: {len(new_questions)}")
    questions.extend(new_questions)
```

# Parse questions to format we use in the app

```python
import random
parsed_questions = []
count = 0
for question in questions:
    options = question["incorrect_answers"].copy()
    options.append(question["correct_answer"])
    if(len(options) != 4):
          raise "options length is not 4" 
    random.shuffle(options)
    questionAndOptions = question["question"] + ", "
    option_indices = ["a", "b", "c", "d"]
    index = 0
    answer = ""
    for option in options:
        questionAndOptions += option_indices[index] + ": " + option
        if(index < len(options)-1):
            questionAndOptions += ", "
        if(option == question["correct_answer"]):
            answer = option_indices[index]
        index += 1
        
    pq = {"id": str(count), "topic": question["category"], "difficulty": question["difficulty"], "questionAndOptions": questionAndOptions, "answer": answer}
    parsed_questions.append(pq)
    
    count +=1
```

# store questions in json

```python
import json 
dictionary = {
"trivia": parsed_questions
}
with open('./generated_questions.json', 'w') as fp:
    json.dump(dictionary, fp)
```
